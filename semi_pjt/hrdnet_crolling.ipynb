```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
from time import sleep
import pandas as pd
#from selenium.common.exceptions import UnexpectedAlertPresentException
data_list = []
for i in range(421873,430000):
            url ='https://www.hrd.go.kr/hrdp/co/pcobo/PCOBO0100P.do?tracseId=AIG20230000'+str(i)+'&tracseTme=1&crseTracseSe=C0061&trainstCstmrId=500020049178#undefined'
            resp = requests.get(url)
            soup = BeautifulSoup(resp.text, 'html.parser')

            title_soup = soup.select_one('.title  h4.tit')
            Is_title = title_soup.get_text(strip=True)
            post_list = soup.select('.content > div.infoList > ul.list')
            print(i,len(Is_title))
            if(len(Is_title) == 0):
                    continue

            span_list = []    

            for post in post_list:
                spans = post.find_all('span')
                for span in spans:
                    #print(span.text.strip())
                    span_list.append(span.text.strip().replace(" ","").replace("\n",""))

            if len(Is_title) > 0:
                data_list.append(tuple([Is_title,span_list[7],span_list[13],span_list[15],span_list[17],span_list[19].replace("",""),span_list[27],span_list[29],span_list[35]]))        
            #print(data_list)

```
