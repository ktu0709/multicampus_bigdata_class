```python
pi = 3.141592 #정적
def cycle_func1(num1): #원 넓이를 구하는 함수
    global pi 
    return pi*num1*num1

def cycle_func2(num1): #원 4/1 부채꼴의 넓이를 구하는 함수
    global pi 
    return 1/4*pi*num1*num1

print(cycle_func1(5))
print(cycle_func2(5))
```

exam01) abs함수를 사용하지 말고 if문을 이용해서 절대값을 출력하는 my_abs라는 함수를 생성해보자

```python
def my_abs(num):
    if num < 0 :
        #return (num - num) - num
        return -1 * num
    else:
        return num

    
print(my_abs(-25))
print(my_abs(-75))
print(my_abs(105))
```

### 가우시안 분포 = 정규분포
- 분산 : 데이터가 평균값으로 부터 퍼져있는 정도의 기대값을 측정한것 ex)var() = 시그마제곱($\sigma^2$)
- 시그마 $\sigma$ (=표준편차) : 데이터의 분포 또는 집단의 표준편차를 나타낸다
- 시그마제곱 : 표준편차를 제곱한 값 = 분산
- 뮤 $\mu$ : 평균값(mean) 또는 기대값
- 정규분포에서 뮤는 분포의 중심이 되는 평균값이다

- 뮤 $\mu$ = 분포의 평균 , 기대값   
- 시그마 $\sigma$ = 표준편차   
- 시그마제곱 $\sigma^2$ = 분산   

### 정규분포인 가우시안 분포일떄 꼬리들
 - 대칭성 : 평균 뮤를 중심으로 좌우 대칭인 종(bell)의 모양을 가진다
 - 평균값 : 평균 뮤는 분포의 중심을 나타내고 , 가장 높은 확률 밀도를 가진다
 - 표준편차 : 시그마는 분포의 쪽을 나타내고 , 값이 클수록 분포는 더 넓게 퍼진다
 - 분산 : 분포의 폭은 시그마제곱에 의해 결정되고 분산이 클수록 데이터는 평균값에서 멀리 떨어진다
 - 68-95-99.7 규칙 : 68% 데이터가 평균에서 표준편차의 1배이내에 , 95%는 2배이내 , 99,7%는 3배이내에 위치한다

 ### 정규분포는 왜?
 - 오차, 측정값, IQ 점수등 데이터분포를 설명
 - 통계적추론 , 신뢰구간 계산 , 가설 검증 등의 기본

exam) 위의 모집단에서 100명 표본을 추출하여 100명의 평균키를 비어있는 리스트 객체 a에 입력하는 작업을  
1000번 수행하여  a 리스트에 10000개의 표본의 평균키가 입력되게 하자

```python
import numpy as np

avg = 148.5
std =30
N = 1000000

mogip = np.random.randn(N) * std + avg

a = []

for i in range(1,10001):
    res = np.random.choice(mogip,100).mean() #모집단으로 부터 100명을 추출하여  평균값
    a.append(res) #평균값을 10000 번함

print(a)
print(len(a))

# 초등학교 100만명의 키 모집단에서 표본을 100개 추출해서 표본평균을 10000개 모았다!!!
```

```python
#pdf() -> 연속확률 변수의 확률 분포를 설명한다
#확률 밀도 함수는 음수가 아니어야 한다 , 적분값(전체확률) = 1
import numpy as np
from scipy.stats import norm

avg = 148.5
std =30
N = 1000000

mogip = np.random.randn(N) * std + avg

a = []

for i in range(1,10001):
    res = np.random.choice(mogip,100).mean() #모집단으로 부터 100명을 추출하여  평균값
    a.append(res) #평균값을 10000 번함

x= np.arange(140,160,0.001) #140~160까지 0.001간격으로 숫자를 생성
y=norm.pdf(x,np.mean(a),np.std(a)) #초등학생 키의 표본평균값들에 대한 확률 밀도함수 값 출력

print(y)
```

```python
'''
pdf() -> 연속확률 변수의 확률 분포를 설명한다
확률 밀도 함수는 음수가 아니어야 한다 , 적분값(전체확률) = 1
정규분포(가우스분포) , 지수분포 , 균일분포 , 베타분포등을 매개변수(평균 , 표준편차, 형상모수, 척도모수)
'''
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

avg = 148.5
std =30
N = 1000000

mogip = np.random.randn(N) * std + avg

a = []

for i in range(1,10001):
    res = np.random.choice(mogip,100).mean() #모집단으로 부터 100명을 추출하여  평균값
    a.append(res) #평균값을 10000 번함

x= np.arange(140,160,0.001) #140~160까지 0.001간격으로 숫자를 생성
y=norm.pdf(x,np.mean(a),np.std(a)) #초등학생 키의 표본평균값들에 대한 확률 밀도함수 값 출력

#print(y)
plt.plot(x,y,color = 'blue')
plt.show()
```
