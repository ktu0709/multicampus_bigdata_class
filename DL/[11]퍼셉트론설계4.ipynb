### subclassing API 작성해보자
1) tf.keras.Model를 상속받는 클래스를 생성
2) 생성자에 레이어 정의
3) call 메소드에 레이어 연결

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import  Sequential
from tensorflow.keras.layers  import  Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses  import mse
import keras
```

```python
import warnings
import tensorflow as tf
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
warnings.filterwarnings('ignore')
```

```python
class my_model(keras.Model):
    
    def __init__(self):
        super(my_model,self).__init__()
        self.dense1 = Dense(units=512,input_shape=(1,), activation='relu')
        self.dense2 = Dense(units=256,activation='relu')
        self.dense3 = Dense(units=2,activation='softmax')
    
    def call(self,x,trainning=False):
        x=self.dense1(x)
        x=self.dense2(x)
        x=self.dense3(x)
        return x
    

model = my_model()
    
```

```python
x_train  = np.random.randint(0,5,(20,1))
y_train = np.where(x_train % 2 ==  0,  [0,1] ,[1,0])


model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train , epochs =200, validation_split=0.2)
```

```python
#tf.keras.Model 상속받는 클래스  생성 -> 생성자에 레이어 정의 -> call 메소드에 레이어 연결
#tf.keras.Layers.Layer 상속받는 클래스  생성 -> 생성자에 레이어 정의 -> call 메소드에 레이어 연결


#1 나만의 relu함수를 만들자
def my_relu(x):
    return tf.math.maximum(0.0,x)

class my_activation(keras.layers.Layer):
    def __init__(self,**kwargs):
        super(my_activation,self).__init__()
    
    def call(self,inputs):
        return my_relu(inputs)

#2 사용자 클래스로 접목


class my_model(keras.Model):
    
    def __init__(self):
        super(my_model,self).__init__()
        
        self.m1 = my_activation()
        self.m2 = my_activation()    
        
        self.dense1 = Dense(units=512,input_shape=(1,))
        self.dense2 = Dense(units=256)
        self.dense3 = Dense(units=2,activation='softmax')
    
    def call(self,x,trainning=False):
        x=self.m1(self.dense1(x))
        x=self.m2(self.dense2(x))
        x=self.dense3(x)
        return x
        
    

model = my_model()
    
```

```python
x_train  = np.random.randint(0,5,(20,1))
y_train = np.where(x_train % 2 ==  0,  [0,1] ,[1,0])


model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train , epochs =200, validation_split=0.2)
```

```python
#Functional API로 계층 레이어를 구현해서 확인하자.

def fun_Block(inputs):
    x = Dense(300,activation='sigmoid')(inputs)
    x = Dense(200,activation='sigmoid')(x)
    outputs = Dense(100,activation='sigmoid')(x)
    return outputs

 
inputs = Input(shape=(28,28)) #1개 
x= Dense(30,activation='sigmoid')(inputs)# 1개 

x= fun_Block(x) # 3개
x= fun_Block(x) # 3개

outputs = Dense(20,activation='sigmoid')(x) # 1개

Model(inputs=inputs, outputs=outputs,name = "fun_Block_API").summary()
```

```python
#Q5) 재귀적 계층 구조를 사용해 보자.  
class My_Block(Layer):
    def __init__(self):
        super().__init__()
        self.dense0 =Dense(units=300, activation="sigmoid", name="dense0")  
        self.dense1 =Dense(units=200, activation="sigmoid", name="dense1")  
        self.dense2 =Dense(units=100, activation="sigmoid", name="dense2")          
    def call(self, inputs):
        x=self.dense0(inputs)
        x=self.dense1(x)
        output = self.dense2(x)
        return output
    
class My_Block_Net(Model):
    def __init__(self):
        super().__init__()
        self.dense0 =Dense(units=30, activation="sigmoid", name="dense0")  
        self.block0 = My_Block() 
        self.block1 = My_Block()        
        self.dense1 =Dense(units=20, activation="sigmoid", name="dense1")        
        
    def call(self, inputs):
        x=self.dense0(inputs)
        x=self.block0(x)
        x=self.block1(x)
        output = self.dense1(x)
        return output    
    
    def my_summary(self, input_shape ):
        my_x  = Input(shape = input_shape , name='my_input')
        my_m   =Model(inputs= my_x, outputs=self.call(my_x), name ='my_test_model')
        my_m.summary() 
        del my_x, my_m
        
model  = My_Block_Net()
model.my_summary((28,28))
```
