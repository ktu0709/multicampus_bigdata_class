```python
#Q3)텐서를 생성하고 속성을 확인 해보자 .
a= tf.constant(2) 
b= tf.constant([1,2]) 
c= tf.constant([[1,2] , [3,4]]) 

print(tf.rank(a)) # 텐서의 랭크를 계산하자.  
print(tf.rank(b))
print(tf.rank(c))
a,b,c

print(c.shape)
```

```python
#Q4)텐서를 생성하고 4칙연산 
a= tf.constant(3) 
b= tf.constant(2) 

print(tf.add(a,b).numpy())
print(tf.subtract(a,b).numpy()   ,type(tf.subtract(a,b).numpy() )  )
print(tf.multiply(a,b).numpy())
print(tf.divide(a,b).numpy()   , type(tf.divide(a,b).numpy()) )
```

```python
#Q5) 넘파이 배열 변환 하고 다시 텐서로 변환 하자. convert_to_tensor()  
c=tf.add(a,b).numpy() # a,b의 합을 numpy 배열 형태로 변환 
c_square = np.square(c,dtype =np.float32) # numpy모듈에 존재하는 제곱 함수를 적용한다.  

c_tensor = tf.convert_to_tensor(c_square) 
print(f'{c} ->  {c_square}  ->    {c_tensor} ') 
type(c_tensor)
```

```python
#Q6)텐서를 생성하고 4칙연산 차원을 만들어서 해보자. 
a= tf.constant([[1,2] ,[3,4] ]  ) 
b= tf.constant([[1,2] ,[3,4] ]  ) 

print(tf.add(a,b).numpy())
print(tf.subtract(a,b).numpy()   ,type(tf.subtract(a,b).numpy() )  )
print(tf.multiply(a,b).numpy())
print(tf.divide(a,b).numpy()   , type(tf.divide(a,b).numpy()) )
```

```python
c=tf.add(a,b).numpy() # a,b의 합을 numpy 배열 형태로 변환 
c_square = np.square(c,dtype =np.float32) # numpy모듈에 존재하는 제곱 함수를 적용한다.  

c_tensor = tf.convert_to_tensor(c_square) 
print(f'{c} \n ->  {c_square} \n ->  {c_tensor} ') 
type(c_tensor)
```

```python
#relu : 음수는 0으로 양수는 그대로 리턴하는 함수
foo = tf.constant([-10, -5, 0.0, 5, 10], dtype = tf.float32)
tf.keras.activations.relu(foo).numpy()
```

```python
#sigmoid : 데이터를 압축해서 리턴 0~1로 압축해서 리턴하는 특징
a = tf.constant([-20, -1.0, 0.0, 1.0, 20], dtype = tf.float32)
b = tf.keras.activations.sigmoid(a)
b.numpy()
```

```python
#softmax : 클래스간의 확률 분포도 출력 -> 다중 클래스 분류 , 출력 총합이 1이 된다
inputs = tf.random.normal(shape=(32, 10))
outputs = tf.keras.activations.softmax(inputs)
tf.reduce_sum(outputs[0,:])
```

```python
#tanh : 출력값을 -1 ~ 1 사이로 압축
a = tf.constant([-3.0, -1.0, 0.0, 1.0, 3.0], dtype = tf.float32)
b = tf.keras.activations.tanh(a)
b.numpy()
```
