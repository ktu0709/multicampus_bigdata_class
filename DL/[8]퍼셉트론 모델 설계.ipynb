#### 단층 , 다층 퍼셉트론 모델 설계 후 결과 확인

```python
#Q1-1) 모듈 임포트
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers.experimental import SGD
from tensorflow.keras.losses import mse
```

```python
#Q1-2) 데이터 준비
data  = np.array([[0,0],[0,1],[1,0],[1,1]])
label = np.array([[0],[1],[1],[1]])

#모델 생성
model = Sequential()


# 모델에 히든 레이어 추가하겠다 -> 2개의 특성을 가진 1차원 데이터를 입력받고 한개의 출력을 가진 Dense층을 만들겠다
#Dense(1,) -> 하나의 히든레이어 만들겠다
#은닉계층이 1개이고 입력계층 단위계수가 2개인것을 [퍼셉트론]이라고 한다
model.add(Dense(1,input_shape=(2,) , activation = 'linear')) #단층 퍼셉트론 -> 퍼셉트론 생성

# 모델 준비 -> 경사하강법 개선에 대한 속성, 로스, 정답률 등 평가지표[정확도, 정밀도, 재현율, F1] 회귀 = 평균절대오차MAE,MSE
model.compile(optimizer=SGD(),loss=mse,metrics = ['acc','Precision','Recall'])

#학습
model.fit(data,label,epochs=200)
```

```python
#평가지표 및 가중치 값 확인 등 정보확인
print(model.get_weights())  #가중치 확인 -> 파일저장[학습결과/모델 재사용 공유/epochs별로 시간간격/앙상블]

print(model.evaluate(data,label))

#예측
pre = model.predict(data)
for a,b in zip(pre,label):
    print(f'예측 : {a},정답 : {b}')
```
