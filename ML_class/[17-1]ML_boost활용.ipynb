```python
# 라이브러리 가져오기

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import model_selection

from sklearn.ensemble import RandomForestClassifier  # 2진 분류 ,(0,1) 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1) 데이터 로드  
np.random.seed(10)
mr  = pd.read_csv('data/mushroom.csv', header = None)
mr.shape 
```

```python
#2) 종속 변수의 데이터 분포  edible식용 = e, Poisonous 독성  =p
mr.iloc[:, 0].value_counts()  #라벨 분포  

#print(type(mr))
#print(type(mr.iterrows()))
#mr.iterrows()
```

```python
print(ord('e'))
print(chr(ord('e')))
```

```python
#3) 데이터의 특징 값이 기호 -> 숫자 변환  

#case 1 : (8124, 23)
data = [] #X
label  =[] #y

for row_index, row  in mr.iterrows():  #8124
    #print(row_index, row)  row_index: 0 ,  row :[    ] <Series>
    label .append (row.iloc[0]) #Y 
    
    row_data = []
    for v in row.iloc[1:] : # 22
        row_data.append(ord(v))  # 문자 -> 아스키 코드 변환  0~ 127 
    data.append(row_data) #데이터 추가 
print(data)
print(label)

#case 2: 
#for j in range(8124):
#    for i in range(23) :
#        mr.iloc[j,i+1] = ord(mr.iloc[j,i+1])
#mr        

```

```python
from sklearn import metrics
#4) 데이터 분할  6093[학습데이터] 2031[검증데이터  ]
X_train, X_test, y_train, y_test = train_test_split(data,label)
print(len(X_train) , len(X_test), len(y_train), len(y_test))

#5) 학습
clf = RandomForestClassifier(n_estimators=100 , random_state =10, n_jobs =-1)
clf.fit(X_train,y_train) 

#6)예측
pred = clf.predict(X_test)

#7)정확도  
ac_score = metrics.accuracy_score(y_test, pred )
print('정확도 : ' ,  ac_score)

#8)리포트 
cl_report = metrics.classification_report(y_test,pred)
print('리포트 \n ', cl_report)
```

```python
#9) fit()실행 결과를 가지고 tree을 이용해서 graphvia를 시각화 하자. 
from sklearn.tree import export_graphviz
#len(clf.estimators_)
estimator=  clf.estimators_[9]
export_graphviz(estimator,out_file='mushroom01.pdf',  class_names  = np.array(['e','p']) ,
                feature_names= list(mr.columns[1:]), 
                filled=True ,rounded =True  
               )
```

#### 11) 평가지표 (2진 분류 ): 정확도, 오차행렬, 정밀도  , 재현율 , F1스코어 , ROC , ACU


```python
# 정확도 평가(평가 지표)
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score,roc_auc_score
from sklearn.metrics import classification_report

def get_clf_eval(y_test,pred):
    accuracy = accuracy_score(y_test,pred)
    print('정확도:',accuracy)

    # 오차 행렬
    confusion = confusion_matrix(y_test,pred)
    print('오차행렬:\n',confusion)
    TN = confusion[0,0]
    FP = confusion[0,1]
    FN = confusion[1,0]
    TP = confusion[1,1]

    # 정밀도(precsion) : 레이블이 반드시 숫자 (0,1)이어야 한다. 
    precision = precision_score(y_test,pred)
    print('정밀도:',precision)
    print('TP/(FP + TP):',TP/(FP + TP))
    
    # 재현율(recall)
    recall = recall_score(y_test,pred)
    print('재현율:',recall)
    print('TP/(FN + TP)',TP/(FN + TP))
    
    # f1_score
    f1 = f1_score(y_test,pred)
    print('F1-스코어:',f1)
    print(2/((1/recall) + (1/precision)))
    print(2*(recall*precision)/(recall+precision))
    
    # roc_auc
    roc_auc = roc_auc_score(y_test,pred)
    print('roc_auc:',roc_auc)
    
```

```python
######12 . 레이블 인코딩 :y_test, pred
######    스케일링 종류   (1,4번이 이상치 영향을 가장 적게 받는다. )
#########  1)StandardScaler (평균 0, 표준1 조정)/ 2)MinMaxScaler(최대값 1:최소값 0)
#3) RobustScaler(4분위:중앙값 0,( 1분위~ 3분위 )= 1 ) MinMaxScaler의 단점 보안 , 분포퍼져라
#4) MaxAbsScaler(0을 기준으로 절대값이 가장 큰수가 1, -1이 되도록 조정한다. )

#주의할 점!!!, Scaler를 하게되면 fit()실행, transform() 적용하는데  test데이터는 
# 실행하면 안된다. !!   fit(), fit_transform() 불가 
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_test_enc = encoder.fit_transform(y_test)#적용하고 실행도 하자. 
print('y_test 인코딩 변환값:',y_test_enc)

pred_enc = encoder.fit_transform(pred)
print('pred 인코딩 변환값:',pred_enc)

get_clf_eval(y_test_enc,pred_enc)
```
