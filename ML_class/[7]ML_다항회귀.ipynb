```python
# matplotlib에서 한글 폰트를 설정하는 방법
import matplotlib.pyplot as plt
plt.rc("font", family="malgun gothic")

# 음수 기호 출력 방법
import matplotlib
matplotlib.rcParams["axes.unicode_minus"]=False

# 그래프를 선명하게 출력하는 방법
%config InlineBackend.figure_format = "retina"

import numpy as np
```

```python
np.random.seed(42)

n_samples = 100
x = 6 * np.random.rand(n_samples) - 3
y = 0.5 * x**2 + x + 2 + np.random.randn(n_samples)

plt.scatter(x, y, edgecolor="black")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x.reshape(-1,1), y)
print(model.score(x.reshape(-1,1), y))
```

```python
pred = model.predict(x.reshape(-1,1))
plt.scatter(x, y, edgecolor="black")
plt.plot(x, pred, color="red")
plt.show()
```

### 다항회귀모델을 생성 및 평가

```python
# 선형 회귀의 가설에 대하여 차수를 높여 분석을 수행합니다.
from sklearn.preprocessing import PolynomialFeatures   # 다항회귀 연산 구현  

pf = PolynomialFeatures(degree=2) #원본 데이터를 다항 특성으로 변경
new_x = pf.fit_transform(x.reshape(-1,1))

model = LinearRegression()
model.fit(new_x, y)

x_data = np.linspace(-3,3,100)
x_poly_data = pf.transform(x_data.reshape(-1,1))
pred = model.predict(x_poly_data)

plt.scatter(x, y, edgecolor="black")
plt.plot(x_data, pred, color="red")
plt.show()
```

```python
print(model.score(new_x, y))
```
