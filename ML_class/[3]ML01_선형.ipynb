### 클러스터

    1) k-means clustering : 데이터 포인트를 비슷한 특성을 가진 여러 그룹(k개의 클러스터)으로 나누는 알고리즘, 클러스터링에서 가장 널리 알려진 기법 중 하나로, 각 클러스터의 중심(centroid)을 기준으로 데이터를 그룹화한다.  클러스터링의 가장 쉬운 기술로 빅 데이터를 분석할 때 자주 사용되는 가장 중요한 기술

    2) k-nearest neighbor(KNN) : KNN은 패턴 인식과 데이터 분류에 자주 사용되는 알고리즘으로, 새로운 데이터 포인트가 주어졌을 때 특징 공간 내에서 가장 가까운 k개의 훈련 데이터 포인트를 기반으로 분류를 수행한다.  지연 학습(lazy learning) 방식에 속하며, 학습 단계가 별도로 없이 분류 시에 모든 계산이 수행된다.  
  
    
###  transformer _변환기 :
데이터 전처리 과정에서 중요한 역할하며 다양한 기술을 통해 데이터의 차원을 축소하거나, 범주형 변수를 숫자형으로 변환하며, 데이터의 스케일을 조정하는 역할을 한다. 모델의 성능을 향상시키고, 더 효과적인 학습과 예측을 가능하게 하는 데 필요하다. 

    - 차원 축소(Dimensionality Reduction): 주성분 분석(PCA), 선형 판별 분석(LDA) 등 
    - 이진화(Binarization): 연속형 데이터를 이진 형태로 변환  
    - 요인분석(Factor Analysis): 관측된 변수들 뒤에 있는 잠재적 요인을 찾아내는 기법  
    - 레이블 인코딩(Label Encoding), 원-핫 인코딩(One-Hot Encoding): 범주형 변수를 숫자형으로 변환   
    - 정규화(Normalizing): 데이터를 특정 범위 내로 스케일링  
    - 척도 조정(Scaling): 데이터의 범위를 조정하여 표준화  
    - 비음수 행렬 분해(Non-Negative Matrix Factorization, NMF): 데이터 행렬을 비음수 요소의 곱으로 분해  
    - 데이터 채우기(Imputing): 결측치를 대체하는 방법  
    - 텍스트 벡터화(Vectorizing): 텍스트 데이터를 수치 벡터 형태로 변환  
    - (Latent Dirichlet Allocation, LDA): 텍스트 문서의 주제를 발견하는 방법  
    - 특이값 분해(Singular Value Decomposition, SVD): 행렬을 특이 벡터와 특이값으로 분해

### Q5)   생각해 보자 

[1차 결론]  :   정답과 예측치의 오차의 단순합을  모두  0으로 만들면  판단이 모호하다. 

[2차 결론]  :  음수의 값으로 변환되는 것이 있기 때문이다. 


[해결책]  :    오차의 결과를 양수로 만들어 주는 작업을 새롭게 하자  
                
                ----------> 오차들의 제곱들의 평균을 구한 후  이 값의 제곱근으로 오차를 정의하자  
                
                
                -------------> 추정된 선으로 부터 점들이 평균적으로 얼마나 떨어져 있는지를 판단한다. 
                
                
                -----------> 오차 제곱의 합을 최소화 하는 회귀계수를 찾자 

RMSE = Root Mean Square Error  
        
$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$  

 ### 비용함수  머신러닝에서 사용하는 용어

Cost Function
$$
f(X) = \beta_1X + \beta_0
$$
$$
Cost(\beta_0, \beta_1) = \sqrt{\frac{1}{n}\sum_{i=1}^{n}{(Y_i - f(X_i))^2}}
$$

<<손실함수(Loss function), 비용함수(Cost function), 목적함수(Objective function)>>  
• 손실함수 : 한 개의 데이터 포인트에서 나온 오차를 최소화하기 위해 정의되는 함수  
• 비용함수 : 모든 오차를 일반적으로 최소화하기 위해 정의되는 함수  
• 목적함수 : 어떠한 값을 최대화 혹은 최소화 시키기 위해 정의되는 함수

```python
def RMSE(a,b, X,y ) :
    RMSE = np.sqrt(  ((y - (a*X+b)) **2).mean()     )
    return RMSE
```

```python
#첫 번째 직선  f(x) = 2x+5
print(RMSE(2,5,X,Y))

#두 번째 직선  f(x) = 30 
print(RMSE(0,30,X,Y))

#세 번째 직선  f(x) = -2x+55 
print(RMSE(-2,55,X,Y))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
def rmse(y_true,y_pred):    
    rmse = np.sqrt(mean_squared_error(y_true,y_pred))
    print('rmse',rmse)
    return rmse

```

```python
#Q3) 비용함수를 이용해서  선형회귀를 확인 해보자. _ 튜터리얼 

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Create linear regression object  -> 객체 생성 
regr = linear_model.LinearRegression()

# Train the model using the training sets  -> 실행  
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set  -> 예측  
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients  회귀계수 
print("Coefficients: \n", regr.coef_)

# The mean squared error   비용함수   
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))

# The coefficient of determination: 1 is perfect prediction  검증 스코어  
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

MSE ( Mean Squared Error ) = 오차 제곱 평균 = 연속형변수값을 예측할 때 사용 : 다중분류시 사용
 
MAE (Mean Absolute Error, 평균 절대 오차)
$$
MAE(\hat{y}) := \frac{1}{N}\sum_{i=1}^{N} \left| y_i - \hat{y}_{i} \right|
$$


\begin{align} 
MSE(\hat{y}_{\theta_1}) &> MSE(\hat{y}_{\theta_2}) \ 
\Leftrightarrow ~~ RMSE(\hat{y}_{\theta_1})   &>   RMSE(\hat{y}_{\theta_2}) \ 
\Leftrightarrow ~~ \hspace{1.5em} {R^{2}}(\hat{y}_{\theta_1})   &<   {R^{2}}(\hat{y}_{\theta_2}) 
\end{align}


 -결정계수(R2)  : Y의 총 변동량 중에서 X에 의해서 설명된 량으로 '표본에 대한 회귀모형의 설명'을 말한다.  
   - 값이 클수록(1에 가까울수록) 모형이 적합하다고 평가할수 있다. 
   - 즉  ,  단순히 Y의 평균값을 사용했을 때 대비 X 정보를 사용함으로써 얻는 성능 향상을 갖는다. 
   - X 변수가 Y 변수의 분산을 얼마나 줄였는가에 대한 정도를 확인 할 때 사용한다. 
   
   - R2 = 1  ->모든 측정값들이 회귀직선에 있을 경우 
   - R2 = 0   ->  X,Y의 관계를 알수 없다.
